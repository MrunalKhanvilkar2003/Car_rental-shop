```python
cars ={
    "car1":{'id':'1','name':'Maruti','Available':'Yes','charges per hour':200},
    "car2":{'id':'2','name':'Ford','Available':'Yes','charges per hour':400},
    "car3":{'id':'3','name':'Hondai','Available':'Yes','charges per hour':600},
    "car4":{'id':'4','name':'Maruti','Available':'yes','charges per hour':800}
    
}
customer_rentals = []
earnings = 0

while True:
    print("Car Rental Shop")
    print("1. Display Available Cars")
    print("2. Rent a Car")
    print("3. Return a Car")
    print("4. Display Earnings")
    print("5. Exit")
    select = input("Select your choice (1/2/3/4/5): ")
    
    if select == "1":
        print("\nAvailable Cars:")
        for car, info in cars.items():
            if info["Available"]=='Yes':
                print(f"{car} ({info['name']}) - Hourly Rate: {info['charges per hour']} Rs")
    elif select == "2":
        car_choice = input("select car you want to rent: ")
        if car_choice in cars and cars[car_choice]["Available"]=='Yes':
            customer_name = input("Enter your name: ")
            rental_hours = int(input("Enter the number of hours you want to rent the car: "))
            if rental_hours <= 0:
                print("Invalid rental duration. Please enter a valid duration.")
                continue

            bill_amount = cars[car_choice]["charges per hour"] * rental_hours
            print(f"Rented {car_choice} for {rental_hours} hours. Your bill is {bill_amount}")
            cars[car_choice]["Available"] = False
            customer_rentals.append({
                "car": car_choice,
                "customer_name": customer_name,
                "rental_hours": rental_hours,
                "bill_amount": bill_amount,
            })
            print(customer_rentals)
            cars[car_choice]['Available']='No'
            earnings=earnings + bill_amount
             
        else:
            print("Sorry, the selected car is not available.")
    elif select == "3":
        if not customer_rentals:
            print("No cars have been rented yet.")
        else:
            return_choice = input("Enter the number of the car you are returning: ")
            if return_choice in customer_rentals[0]['car']:
                    customer_rentals.pop(0)
                    print(f"Thank you for returning the {return_choice}. Your total rental fee is ${bill_amount}.")
                    cars[ return_choice]['Available']='True'
            else:
                    print("Sorry, we couldn't find a record of this car being rented from us.")
    elif select== "4":
        print(f"Total Earnings: {earnings} Rs")
      
    elif select == "5":
        break

    else:
        print("Invalid select. Please select a valid option.")
```
